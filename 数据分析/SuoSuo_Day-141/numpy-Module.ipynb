- 二维数组的创建

```python
np.array([[1,2,3,'wan'],['fd',12,1.1]])  # 不规则的 二维数组
```

```python
np.array([[1,2,3,],['fd',12,1.1]])      # 规则的二维数组，每行的列数相同
```

注意:
    - 列表可以存放 不同的数据结构， 是因为 开辟的内存空间中存放的数据实际是 不同数据 的内存地址
- numpy 默认 nbarray 的所有元素的类型是相同的
    - 而 数组不一样， 在开辟的内存空间中存放的是真实的 数据
    -     - 如果传进来的列表中包含不同的数据，则统一为 同一类型， 优先级: str>float>int

##  ndarray的属性

4个必记参数：
ndim：维度
shape：形状（各维度的长度）
size：总长度

dtype：元素类型

```python
img_arr.shape  # 三个数代表三个维度， 前两个代表 像素为: 320*240,; 第三个数代表每个像素点的颜色值
```

```python
img_arr.ndim  # 返回此数组的 维度数，3 代表 有 三个 维度
```

```python
img_arr.size  # 返回此数组中 一共 有多少个数组元素
```

```python
img_arr.dtype # 返回此数组中的数组元素的类型，uint8：代表每一个数组元素占的内存为 8 位
```

```python
type(img_arr) # 返回此数组的 类型
```
