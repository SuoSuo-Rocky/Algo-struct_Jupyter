```python
# 爬取煎蛋网 http://jandan.net/ooxx  图片  # 反爬机制 ： 数据的加密  base64 

## 注意 此次爬取的 煎蛋网 图片 基于 以前的 煎蛋网 源码数据， 
# 现在的煎蛋网 源码数据 已经将 图片的 加密 取消 了
import requests
from lxml import etree
import urllib
import os
import base64
```

```python
#  以前的  
url = 'http://jandan.net/ooxx'
header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; rv:66.0) Gecko/20100101 Firefox/66.0',
}
response = requests.get(url=url,headers=header)
pg_text = response.text

if not os.path.exists('./jiandan'):
    os.mkdir('./jiandan')

tree = etree.HTML(pg_text)
img_hash_list = tree.xpath('//span[@class="img-hash"]/text()')

for hash in img_hash_list:
    img_url ="http:"+base64.b64decode(hash).decode()         # 反反爬机制， 将加密的数据 解密 ，
    img_name ='./jiandan/'+ img_url.split('/')[-1]
    urllib.request.urlretrievee(url=img_url,filename=img_name)
    print(img_name + '下载成功！')
print('over')
```
